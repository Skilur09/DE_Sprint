Какая средняя и медианная зарплата по группам вакансий?

```python
df.groupby('job_title')['salary_in_usd'].mean()
```

```python
df.groupby('job_title')['salary_in_usd'].median()
```

Какая средняя и медианная зарплата по каждому региону?

```python
df.groupby('company_location')['salary_in_usd'].mean().round(2)
```

```python
df.groupby('company_location')['salary_in_usd'].median()
```

Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
d=df.groupby('job_title')['salary_in_usd']
avg=pd.DataFrame(d.sum()/d.count())
avg.sort_values(['salary_in_usd'], ascending=False).iloc[0]
```
